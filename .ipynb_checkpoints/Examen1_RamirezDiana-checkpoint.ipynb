## I. Primera parte (40%)

**Pregunta 1.** Con base en la siguiente distribución de rendimientos para el activo A, ¿cuál es su desviación estándar?

| Probabilidad | Rendimiento |
| ------------ | ----------- |
| 0.3          | 0.10        |
| 0.4          | 0.05        |
| 0.3          | 0.30        |

A. 12.15%

B. 8.34%

C. 10.68%

D. 15.42%

La respuesta correcta es (4%): C

```python
# La justificación a esta pregunta son los cálculos necesarios para llegar al resultado (4%)
import pandas as pd
tabla3 = pd.DataFrame(columns=['ret'], index=[0.3, 0.4, 0.3])
tabla3.index.name = 'prob'
tabla3['ret']=[.1,.05,.3]
import numpy as np
E_r = sum(tabla3.index*tabla3.ret)
sqrt_r = np.sqrt(sum(tabla3.index*(tabla3.ret-E_r)**2))
sqrt_r
```

Además, suponga que va a tener un rendimiento anual en sus inversiones del 10% en caso de recesión, 20% en caso de normalidad y 40% en caso de expansión. ¿Cuál sería la desviación estándar del rendimiento anual de sus inversiones?

**Pregunta 3.** Considerela siguiente distribución de rendimientos de los activos A, B y C:


| Probabilidad     | Rendimiento A       | Rendimiento B         | Rendimiento C         |
| ---------------- | ------------------  | -------------------   | ------------------    |
| 30%              | -0.20               | -0.05                 | 0.05                  |
| 40%              | 0.05                | 0.10                  | 0.03                  |
| 30%              | 0.40                | 0.15                  | 0.02                  |


Calcule el rendimiento esperado de un portafolio igualmente ponderado por los activos A, B y C.

A. 4.89%

B. 9.32%

C. 6.10%

D. 5.32%

La respuesta correcta es (4%): C

```python
# La justificación a esta pregunta son los cálculos necesarios para llegar al resultado (4%)
tabla = pd.DataFrame(columns=['Prob', 'A', 'B', 'C'], index=['Expansion', 'Normal', 'Recesion'])
tabla.index.name = 'State'
tabla['Prob']=np.array([0.3, 0.4, 0.3])
tabla['A']=np.array([-0.2, 0.05, 0.4])
tabla['B']=np.array([-.05,.1,.15])
tabla['C']=np.array([0.05,.03,.02])
tabla.loc['Rend_esp']= np.array([None,
                                np.sum(tabla.Prob*tabla.A),
                                np.sum(tabla.Prob*tabla.B),
                                np.sum(tabla.Prob*tabla.C)])
tabla['PortABC']=(1/3)*tabla['A']+(1/3)*tabla['C']+(1/3)*tabla['B']
tabla
```

## II. Segunda parte (60%)

**Punto 1 (20%).** En el archivo `data.csv` se encuentran reportados los precios mensuales ajustados en el cierre de los activos de Microsoft, Grupo Financiero Banorte, Grupo Financiero Inbursa, Cemex y Grupo Carso.

1. Importar los precios de *Microsoft (MSFT) y de Grupo Carso (GCARSOA1.MX)* en un DataFrame indizado por fecha. Mostrar la gráfica de estos precios (5%).
2. Obtener un DataFrame con los rendimientos porcentuales mensuales para estos activos. Mostrar la gráfica de rendimientos (5%).
3. Reportar en un DataFrame el rendimiento esperado mensual y la volatilidad mensual para cada uno de estos activos (5%).
4. Calcular las matrices de varianza-covarianza (base mensual) y de correlación (base mensual) para estos activos (5%).

```python
closes=pd.read_csv('data.csv',index_col='Date',usecols=['Date','MSFT','GCARSOA1.MX'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
closes.plot(figsize=(8,6))
```

```python
ret=closes.pct_change().dropna()
ret
```

```python
ret.plot(figsize=(8,6));
```

```python
ret_sum=pd.DataFrame(index=['Media','Volatilidad'],columns=closes.columns)
ret_sum.loc['Media']=ret.mean()
ret_sum.loc['Volatilidad']=ret.std()
ret_sum
```

```python
cov=ret.cov()
cov
```

```python
corr=ret.corr()
corr
```

**Punto 2 (15%).** Usando los datos del punto anterior:

1. Construya la frontera de mínima varianza que representa las posibles combinaciones de portafolios que se pueden crear con los activos del primer punto. Reportar los rendimientos esperados y las volatilidades de estos portafolios en un DataFrame (8%).
2. Encontrar los pesos del portafolio de mínima varianza. ¿Cuáles son el rendimiento esperado y volatilidad del portafolio de mínima varianza? (7%).

```python
# Vector de w variando entre 0 y 1 con n pasos
w = np.linspace(0, 1, 40)
# Rendimientos esperados individuales
# Activo1: msft, Activo2:carso
E1 = ret_sum['MSFT']['Media']
E2 = ret_sum['GCARSOA1.MX']['Media']
# Volatilidades individuales
s1 = ret_sum['MSFT']['Volatilidad']
s2 = ret_sum['GCARSOA1.MX']['Volatilidad']
# Correlacion
r12 = corr['MSFT']['GCARSOA1.MX']
portafolios1 = pd.DataFrame(index=w,columns=['Rend','Vol'])
portafolios1.index.name = 'w'
portafolios1.Rend = w*E1+(1-w)*E2
portafolios1.Vol = np.sqrt((w*s1)**2+((1-w)*s2)**2+2*w*(1-w)*r12*s1*s2)
portafolios1.round(4)
```

```python
minvar = (s2**2-(r12*s1*s2))/(s1**2+s2**2-2*(r12*s1*s2))
w=round(minvar,6)
print(w*100,'% en MSFT, ',(1-w)*100,'% en GCARSOA1.MX')
```

```python
#para comprobar
import scipy.optimize as opt
def var2(w,s1,s2,s12):
    return (w*s1)**2+((1-w)*s2)**2+2*w*(1-w)*s12
w0=0
# Volatilidades individuales y covarianza
s12 = corr['MSFT']['GCARSOA1.MX']*s1*s2

# Cota de w
bnd=(0,1)
minvar=opt.minimize(var2,w0,args=(s1,s2,s12),bounds=(bnd,))
minvar
```

```python
er=w*E1+(1-w)*E2
print('Rendimiento esperado: ',er)
vol = np.sqrt((w*s1)**2+((1-w)*s2)**2+2*w*(1-w)*r12*s1*s2)
print('Volatilidad: ',vol)
```

**Punto 3 (15%).** En este punto se usarán los mismos activos con sus rendimientos esperados y volatilidades, solo que supondremos que el coeficiente de correlación entre sus rendimientos se supondrá de 0.5.

1. Nuevamente, construya la frontera de mínima varianza para estos dos activos con el nuevo supuesto de correlación. Reportar los rendimientos esperados y las volatilidades de estos portafolios en un DataFrame (8%).
2. Encontrar los pesos del portafolio de mínima varianza. ¿Cuáles son el rendimiento esperado y volatilidad del portafolio de mínima varianza? (7%).

```python
# Vector de w variando entre 0 y 1 con n pasos
w = np.linspace(0, 1, 40)
# Rendimientos esperados individuales
# Activo1: msft, Activo2:carso
e1 = ret_sum['MSFT']['Media']
e2 = ret_sum['GCARSOA1.MX']['Media']
# Volatilidades individuales
S1 = ret_sum['MSFT']['Volatilidad']
S2 = ret_sum['GCARSOA1.MX']['Volatilidad']
# Correlacion
R12 = .5
portafolios2 = pd.DataFrame(index=w,columns=['Rend','Vol'])
portafolios2.index.name = 'w'
portafolios2.Rend = w*e1+(1-w)*e2
portafolios2.Vol = np.sqrt((w*S1)**2+((1-w)*S2)**2+2*w*(1-w)*R12*S1*S2)
portafolios2.round(4)
```

```python
minvar2 = (S2**2-(R12*S1*S2))/(S1**2+S2**2-2*(R12*S1*S2))
w=round(minvar2,6)
print(w*100,'% en MSFT, ',(1-w)*100,'% en GCARSOA1.MX')
```

```python
#para comprobar
import scipy.optimize as opt
def var2(w,s1,s2,s12):
    return (w*s1)**2+((1-w)*s2)**2+2*w*(1-w)*s12
w0=0
# Volatilidades individuales y covarianza
S12 = .5*S1*S2

# Cota de w
bnd=(0,1)
minvar2=opt.minimize(var2,w0,args=(S1,S2,S12),bounds=(bnd,))
minvar2
```

```python
Er=w*e1+(1-w)*e2
print('Rendimiento esperado: ',Er)
vol = np.sqrt((w*S1)**2+((1-w)*S2)**2+2*w*(1-w)*R12*S1*S2)
print('Volatilidad: ',vol)
```

```python
plt.figure(figsize=(8,6))
plt.plot(s1, E1, 'ro', ms = 10, label='MSFT')
plt.plot(s2, E2, 'bo', ms = 10, label='GCARSOA1.MX')
plt.plot(portafolios1.Vol, portafolios1.Rend, 'g-', lw = 4, label='Portafolios1')
plt.plot(portafolios2.Vol, portafolios2.Rend, 'k-', lw = 4, label='Portafolios2')
plt.plot(np.sqrt(minvar.fun),minvar.x*E1+(1-minvar.x)*E2,'*m',ms=10,label='Port.min.var1')
plt.plot(np.sqrt(minvar2.fun),minvar2.x*e1+(1-minvar2.x)*e2,'*b',ms=10,label='Port.min.var2')
plt.title('Volatilidad vs rendimiento')
plt.xlabel('Volatilidad ($sigma$)')
plt.ylabel('Rendimiento esperado ($E[r]$)')
plt.legend(loc='best')
plt.grid()
plt.show()
```
