## Descripción del proyecto

# Resumen

El presente proyecto se enfoca en la utilización de un método de análisis de estilo, específicamente el análisis de estilo basado en retornos, para lograr obtener, de un portafolio, las ponderaciones de estilos de inversión basados estrategias comunes como lo son los diversos índices de mercado existentes. 

Los objetivos centrales de este proyecto se enfocaron en identificar la ponderación de estilo de mercado ($\beta$) y de elección discrecional sobre un portafolio ($\alpha$). Así como obtener los rendimientos de índices de cada estilo seleccionado, tanto como el de un portafolio público. En este caso el portafolio público tomado fue el **T. Rowe Price Blue Chip Growth Fund (TRBCX)** de "Yahoo Finance". En el caso de los índices, se tomaron los siguientes de la página "S&P Down Jones Indices" desde el año 2008 hasta el mes de noviembre del 2018. Posteriormente se obtuvieron sus rendimientos: ***S&P 500 Growth, S&P 500 Value, S&P Midcap 400 Growth, S&P Midcap 400 Value, S&P 600 SmCap Growth, S&P 600 SmCap Value , y el Treasury Bill.*** 

Después de esto, se planteó una solución utilizando las herramientas del lenguaje de programación Python y la plataforma Jupyter para manejar los datos de rendimientos y mostrarlos en presentaciones de tabla, y gráfica. Posteriormente, se optimizó la fórmula de "Estilo de Análisis basada en rendimientos" dando como resultado la ponderación correspondiente de las variables descriptivas del portafolio. El portafolio queda representado en un 84.62% por el S&P 500 Growth, 9.19% por el S&P 600 Growth, y 6.19% por el S&P 400 Growth. Posteriormente, se representó en una gráfica la solución correspondiente del análisis de varianza de nuestro modelo, y se calculó R, mediante este análisis y mediante mínimos cuadrados, resultando en 94.98%. Con esta gráfica pudimos visualizar al modelo aproximandose a casi nula diferencia con el rendimiento actual del fondo elegido. 
Se considera que era esperada esta relación con el ***S&P 500 Growth***  dado que el portafolio seleccionado utiliza explícitamente una estrategia Growth (alta capitalización).

# Antecedentes

   ### Análisis de Estilo


  - **El análisis de estilo** es un proceso para determinar cual es el comportamiento de inversión que un administrador tiene al hacer sus decisiones de inversión. Siendo estas evaluadas a un estilo de inversión específico. Hay un número ilimitado de estilos de inversión, sin embargo, los tipos más comunes son:  _growth, value, large-cap, y small-cap_. (Investopedia, 2018). 
        
        
  - El método utilizado se basa en una téncica estadística para deconstruir los retornos de inversión, dados diferentes variables explicativas. El modelo resultante muestra las clases de activos a los que una estrategia esta expuesta. Generalmente en ejes comunes de estilo, como índices. Esta medida se puede utilizar en fondos o para determinar estilos que los manejadores de portafolios usan para generar sus retornos.
        
        
  - El modelo fue **elaborado en 1989 por William F. Sharpe**, en su artículo _"Determining a Fund's Effective Asset Mix"_. Y todavia hasta el día de hoy es utilizado en análisis de fondos. Los modelos de aplicación se centran en repetidas regresiones que calculan un **alpha** y los **vectores de beta** para cada variable. Resultando en una imagen estadística que demuestra el estilo de un manejador de dinero comparado a estrategias similares y más generales del mercado (Sharpe, 1988).
  
  ![](https://wikimedia.org/api/rest_v1/media/math/render/svg/3d56f73f83cc47e1a344f1aa0d875eb46e15a243)
  
  
  
   - El _**"análisis de estilo basado en retornos"**_ se usa preferencialmente, puesto que los datos mensuales de retorno están altamente disponibles para cualquier clase de activo. Por lo que se facilita su manejo y selección de un portafolio para el inversionista. Aunque tambien existe el **_"análisis de estilo basado en valor de cartera"_**. Por ejemplo, un _"análisis de estilo basado en valor de cartera"_ se enfocaría más al análisis del tamaño y valor , así como la orientación, de las acciones subyacentes en un fondo. A diferencia de un _"análisis de estilo basado en retornos"_ que compara al fondo y sus retornos totales contra los retornos totales de varios índices, y hace inferencias sobre qué tan cerca están los retornos de este portafolio a los de diferentes índices (MorningStart, 2007).
        

# Fondos utilizados en el problema

- ### S&P 500

    - S&P 500 Growth
        
        - La parte del S&P 500 que se considera que tiene "características de crecimiento" constituye el índice de crecimiento de S&P 500. Si bien no hay cortes específicos para distinguir las acciones de crecimiento, generalmente son (pero no siempre) empresas con un crecimiento de ventas superior al promedio, y se intercambian por relaciones de precio a ganancias relativamente altas. Para darle una idea, algunas de las acciones más grandes en el índice de crecimiento S&P 500 son Apple, Amazon, Facebook, Visa y Boeing.

    - S&P 500 Value
    
        - S&P 500 Value Index consiste en las acciones en el S&P 500 que se considera que tienen "características de valor". Generalmente, se trata de acciones que se negocian por múltiplos relativamente bajos de sus valores contables y ganancias, y tienden a ser compañías más maduras y de crecimiento más lento. Algunas de las acciones más grandes en el Índice de valor S&P 500 incluyen JPMorgan Chase, Berkshire Hathaway, AT&T y ExxonMobil.  (Matthew Frankel, 2018)

    
- ### S&P Midcap 400

    - S&P Midcap 400 Growth
    
         - La inversión busca rastrear el desempeño de un índice de referencia que mide el retorno de la inversión de las acciones de crecimiento de capitalización media en los Estados Unidos. El fondo emplea un enfoque de inversión de indexación diseñado para rastrear el desempeño del Índice de Crecimiento S&P MidCap 400®, que representa a las compañías en crecimiento, según lo determina el patrocinador del índice, del Índice S&P MidCap 400. El Asesor intenta replicar el índice objetivo invirtiendo todos, o sustancialmente todos, de sus activos en las acciones que componen el índice, manteniendo cada acción en aproximadamente la misma proporción que su ponderación en el índice.
        
    - S&P Midcap 400 Value
        
         - Invierte en acciones en el S&P MidCap 400 Value Index, compuesto por las empresas de valor en el S&P 400. Se enfoca en seguir de cerca el rendimiento del índice, que se considera un indicador de los rendimientos de las acciones de valores de capitalización media en los EE. UU. Ofrece un alto potencial para el crecimiento de la inversión; el valor de las acciones sube y baja más bruscamente que el de los fondos que tienen bonos. Más apropiado para objetivos a largo plazo donde el crecimiento de su dinero es esencial.

 

 

  
    
- ### S&P 600 SmCap 

    - S&P 600 SmCap Growth
    
        - El Índice de Crecimiento S&P SmallCap 600 representa a las compañías en crecimiento del Índice S&P SmallCap 600. (El índice S&P SmallCap 600 representa el segmento de pequeña capitalización del mercado de valores de EE. UU.). Realiza un seguimiento de las empresas en crecimiento de este índice identificadas por tres factores: tasa de crecimiento de las ganancias por acción a tres años, tasa de crecimiento de las ventas por acción a tres años y momento (cambio de 12 meses en el precio)
         
    - S&P 600 SmCap Value
    
        - Este fondo de índice de bajo costo ofrece exposición a las acciones de pequeña capitalización de valores de Estados Unidos. Las acciones de valor son aquellas que pueden ser infravaloradas temporalmente por los inversores. Estas empresas suelen crecer a un ritmo más lento que la empresa típica. El fondo busca rastrear un índice de estilo de valor de pequeñas empresas. Uno de los principales riesgos del fondo es su enfoque en la arena de pequeña capitalización, que es un segmento del mercado a menudo volátil. Los inversores que deseen agregar una asignación de valor de pequeña capitalización administrada pasivamente a una cartera ya diversificada pueden considerar este fondo.
    
    
#### Treasury Bill

##### Definición

Son bonos del Estado o títulos de deuda con vencimiento inferior a un año. 

##### Descripción

Las facturas en T se emiten para cubrir los desajustes a corto plazo en los recibos y gastos. Los bonos de mayor vencimiento se denominan valores fechados.  (Bennett, 2018)



#### Portafolio público utilizado

- T. Rowe Price Blue Chip Growth Fund (TRBCX)

##### Objetivo de Inversión

El objetivo del fondo es proporcionar un crecimiento del capital a largo plazo. El ingreso es un objetivo secundario.

##### Estrategia

El fondo invierte principalmente en acciones ordinarias de grandes y medianas empresas Blue Chip que tienen potencial para un crecimiento de ganancias superior a la media y están bien establecidas en sus respectivas industrias.

Este fondo ofrece una apreciación del capital a largo plazo mediante la inversión en empresas líderes en industrias que T. Rowe Price cree que están preparadas para el crecimiento a largo plazo. Brinda a los inversionistas una forma de bajo costo para invertir en compañías "blue chip" que disfrutan de posiciones sólidas en el mercado, equipos de administración experimentados, condiciones financieras sólidas y un crecimiento y rentabilidad de las ganancias por encima del promedio. 

Debido a que las acciones de crecimiento tienen valoraciones más altas y rendimientos de dividendos más bajos que las compañías de crecimiento más lento o cíclicas, la volatilidad del precio de las acciones puede ser mayor. Como tal, los precios de los fondos podrían disminuir aún más en las caídas del mercado que los fondos no orientados al crecimiento. (T. Rowe Price, 2018)



## Solución a la problemática planteada

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
from scipy.optimize import minimize
```

```python
data = pd.read_excel('Datos_indices.xlsx', index_col='Effective date')
data
```

```python
data.plot(figsize=(8,6));
```

```python
daily_ret_index = data.pct_change().dropna()
daily_ret_index
```

```python
daily_ret_index.plot(figsize=(8,6));
```

```python
# Importamos scipy
from scipy.optimize import minimize
# Construir matriz I
I = np.concatenate((np.ones((len(daily_ret_index.index), 1)),np.array(daily_ret_index.iloc[:,0:7])), axis=1)
# Construimos vector R
R = np.array(daily_ret_index.iloc[:,7])
```

```python
# Función objetivo
def error(aw, R, I):
    return np.linalg.norm(R-I.dot(aw))**2
```

```python
# Dato inicial
aw0 = np.ones((8,))/8
# Cotas de las variables
bnds = ((None,None),)+((0,1),)*7
# Restricciones
cons = ({'type': 'eq', 'fun': lambda aw: np.sum(aw[1:])-1},)
# Portafolio de mínima varianza
#min_var = opt.minimize(varianza, w0, args=(Sigma,), bounds=bnds, constraints=cons)
min_error = minimize(error, aw0, args=(R,I), bounds=bnds ,constraints=cons)
min_error
```

```python
columns = data.columns.drop('Blue Chip').insert(0,'α')
pesos1 = min_error.x
pesos =pd.DataFrame(min_error.x, index=columns).transpose()
pesos
```

```python
e = R-I.dot(min_error.x)
e
```

```python
#Calculamos R
r = 1-np.var(e)/np.var(R)
r
```

## Visualización de los resultados 

```python
pesos
```

```python
rt_nue = pesos1.dot(I.transpose())
rt_nue = rt_nue.transpose()
```

```python
resultados = pd.DataFrame(index=daily_ret_index.index, columns=['rt','rt_nueva','Errores',  'st'])
resultados.rt = daily_ret_index['Blue Chip']
resultados.rt_nueva = rt_nue
resultados.st = (resultados.rt.mean()-resultados.rt)**2
resultados.Errores = ((resultados.rt-resultados.rt_nueva)**2).round(6)
resultados
```

```python
erre = (np.sum(resultados.st)-np.sum(resultados.Errores))/np.sum(resultados.st)
erre
```

```python
plt.figure(figsize=(50,15))
plt.plot(daily_ret_index.index, resultados.rt)
plt.plot(daily_ret_index.index, resultados.rt_nueva)
plt.xlabel("Tiempo (año)", fontsize =35)
plt.ylabel("Rendimiento", fontsize =35)
plt.title('Comparación real vs calculado',fontsize=40)
plt.rc('axes', titlesize=20)
plt.show()
```
